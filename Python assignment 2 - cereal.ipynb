```python
# Task 4: Using mfr: Manufacturer of cereal columns find the number of cereal per manufacturer

#num_of_cereals = df.groupby('mfr').nunique()
num_of_cereals = df['mfr'].value_counts()
# Print the result
print("Number of cereals per manufacturer:")
print(num_of_cereals)
```

```python
#  Visualize it using count plot and pie chart.
sns.countplot(x = 'mfr', data = df)
```

```python
# pie chart
plt.pie(df['mfr'].value_counts(),labels=['K','G','P','Q','R','N','A'], autopct ='%1.2f%%')
plt.title('Pie chart for outcome')
```

```python
# Task 5: Create a Histogram for the distribution of calories
df.calories.hist(figsize=(5,5)) 
#df.hist(calories) 
  
# Set title 
plt.title("The distribution of calories") 
  
# adding labels 
#df.plot(x ='name',y = 'calories', kind ='hist')
```

```python
# Task 6:Visualize the distribution of 'vitamins' content in cereals. You can use a histogram for this
df.vitamins.hist(figsize=(5,5))  
  
# Set title 
plt.title("The distribution of vitamins") 
```

```python
# Task 7: Find the correlation matrix
col=['calories','protein','fat','sodium','fiber','carbo','sugars','potass','vitamins','shelf','weight','cups','rating']
corr_matrix = df[col].corr()
corr_matrix
```

```python
# Create a correlation heatmap.
col=['calories','protein','fat','sodium','fiber','carbo','sugars','potass','vitamins','shelf','weight','cups','rating']
corr_heat = df[col].corr()
plt.figure(figsize=(10,10))
sns.heatmap(corr_heat, annot=True)
```

```python
# Task 8: Find the correlation between Calories vs Protein.
corr_cal_vs_prot = df['calories'].corr(df['protein'])
corr_cal_vs_prot
```

```python
# Visualize this using scatter plot.
scat_cal_vs_prot = df.plot(x ='calories',y = 'protein', kind ='scatter')
scat_cal_vs_prot
```
